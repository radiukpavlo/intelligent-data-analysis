**Ви будете робити наступне.**

1. **Контентні рекомендації (Content-Based):** cтворите системи, які пропонують щось на основі характеристик елементів – наприклад, жанрів фільмів або тегів товарів. Ви перевірите їхню якість, оцінивши, наскільки ці пропозиції логічні.  
2. **Колаборативна фільтрація (Collaborative Filtering):** зробите системи, які використовують вподобання користувачів (оцінки, прослуховування, покупки) для формування пропозицій. Ви протестуєте їх за допомогою таких метрик, як `RMSE`, `MAE`, `precision@k` та `recall@k`.  
3. **Факторизація матриць (Matrix Factorization):** створите системи, які розкладають дані на патерни (використовуючи такі методи, як SVD або ALS), щоб вгадати, що може сподобатися користувачам. Протестуєте їх за допомогою `RMSE`, `MAE`, `precision@k`, `recall@k`, `MAP` та `NDCG@k`.  
4. **Прийоми ранжирування та оцінювання:** випробуєте способи ранжирування пропозицій – наприклад, поєднання популярності з інтелектуальними прогнозами. Перевірите, наскільки вони ефективні, використовуючи `precision@k`, `recall@k`, `MAP` та `NDCG@k`.  
5. **Гібридні та ансамблеві системи:** поєднаєте ідеї контентної та колаборативної фільтрації або скомбінуєте кілька систем (наприклад, за допомогою стекінгу або усереднення). Перевірите, чи працюють вони краще за поодинокі системи, використовуючи `RMSE`, `MAE`, `precision@k`, `recall@k`, `MAP` та `NDCG@k`.  

**Як ми будемо вимірювати успіх:**

Ви використовуватимете ці метрики, щоб оцінити роботу ваших систем:

- **Метрики регресії:** `RMSE` та `MAE` (для прогнозування оцінок).  
- **Метрики ранжирування:** `Precision@k`, `Recall@k`, `MAP` та `NDCG@k` (для відбору найкращих рекомендацій).

**Варіант 1 – MovieLens: контентне зіставлення фільмів за жанрами та роком**

- **Мета:** побудувати content-based рекомендатор, який знаходить фільми, подібні до `Toy Story (1995)`, на основі жанрового профілю та року випуску.
- **Кроки:**
    1. Завантажте таблиці з фільмами та оцінками, виокремте записи про жанри, назви та роки випуску, перевірте коректність розбору року з назви фільму.
    2. Побудуйте векторне представлення контенту: закодуйте жанри як багатозначні ознаки, рік випуску нормалізуйте та за потреби зменште його вплив окремим коефіцієнтом.
    3. Обчисліть cosine similarity між `Toy Story (1995)` та всіма іншими фільмами, вилучіть сам фільм із вибірки та сформуйте список `top-N`.
    4. Оцініть рекомендації якісно: перевірте жанрову спорідненість, близькість року випуску та відповідність очікуванням аудиторії сімейної анімації.
    5. Подайте таблицю рекомендованих фільмів, візуалізацію подібності та стислий висновок про сильні й слабкі сторони контентного підходу.
- **Підказки:** для жанрів доцільно використати `str.get_dummies('|')` або `CountVectorizer`; рік випуску варто стандартизувати перед об’єднанням з текстовими ознаками.

**Варіант 3 – Jester: контентне наближення жартів за текстом**

- **Мета:** реалізувати рекомендацію жартів на основі текстової подібності з використанням TF-IDF.
- **Кроки:**
    1. Завантажте тексти жартів і, якщо потрібно, таблицю оцінок; очистьте текст від службових символів, HTML-розмітки та порожніх записів.
    2. Побудуйте матрицю TF-IDF для корпусу жартів, експериментально доберіть `ngram_range`, мінімальну частоту термінів та обробку стоп-слів.
    3. Виберіть один опорний жарт, обчисліть cosine similarity між його вектором і всіма іншими жартами, побудуйте `top-N` рекомендацій.
    4. Перевірте, чи мають рекомендовані жарти близьку тематику, форму подачі або лексичний стиль; зафіксуйте приклади як вдалих, так і сумнівних збігів.
    5. Подайте висновок про те, наскільки чисто текстова подібність придатна для моделювання гумору без урахування реакції користувачів.
- **Підказки:** для коротких текстів часто працюють біграми та `sublinear_tf=True`; надто агресивне очищення може прибрати стилістичні маркери жарту.

- **Кроки:**
    1. Реалізуйте контентну модель за тегами треків і виконавців та окремо факторизаційну модель для play counts.
    2. Нормалізуйте обидва джерела скорів і побудуйте кілька вагових конфігурацій гібриду.
    3. Для однакової вибірки користувачів сформуйте `top-N` рекомендацій для кожної конфігурації.
    4. Оцініть модель за `NDCG@k` і `precision@k`, виберіть найкраще поєднання ваг.
    5. Поясніть, чи допомагає гібрид зменшити ефект популярності та покращити персоналізованість рекомендацій на великому каталозі.
- **Підказки:** при порівнянні ваг контролюйте однаковий набір тестових користувачів; контентний сигнал особливо корисний для треків із недостатньою історією прослуховувань.
