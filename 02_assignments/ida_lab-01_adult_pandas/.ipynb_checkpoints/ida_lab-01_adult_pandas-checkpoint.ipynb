```python
male = data[data['sex'] == 'Male']
married_mens = male[male['marital-status'].str.startswith('Married')]
unmarried_mens = male[~male['marital-status'].str.startswith('Married')]
median_age_married = married_mens.groupby('race')['age'].median()
median_age_unmarried = unmarried_mens.groupby('race')['age'].median()

print("Медіанний вік одружених чоловіків за расою:", median_age_married)
print("\nМедіанний вік неодружених чоловіків за расою:", median_age_unmarried)
```
